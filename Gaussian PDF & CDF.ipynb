```python
import numpy as np
import pandas as pd 
from scipy.stats import norm
import matplotlib.pyplot as plt 

#Standard Normal/Normal PDF (PDF)
def find_pdf_value(x,mu,std):
    return scipy.stats.norm.pdf(x,loc=mu,scale=std)


print(f'Standard Normal Evaluated at f(.5) : {find_pdf_value(.5,0,1)}') #Standard Normal
print(f'Normal PDF: Mu = 1.33, Sigma = 4, Evaluated at f(3) : {find_pdf_value(3,1.33,4)}') #Random Normal Variable
```

## PDF Evalauted at a Point
The Probability Density Function of the standard normal/normal distribution represents the frequency of an event occuring given a value at X. X represents the mapping of a random variable to the real number line. The PDF distribution is dependent on the Z-Score (x-mu)/sigma which is taken as the e^(z-score) in the PDF function. For a standard normal variable the mu is 0 and sigma =1 so the Z-Score is simply X. This is different for the random normal variable evaluated at 3 and mu = 1.33 and sigma = 4. We cannot say that these are probabilities, because the integral of a single point does not exist. Rather we use the CDF to determine the probabilities.

```python
values_to_compute_std_norm = np.arange(-4,4,.10) #Create an array of x-values to compute the y-values
values_to_compute_norm = np.arange(-10,11,.10) #Need different array size because mu for this one is 1.33. This way the plot will show the right hand side fully

std_norm = []
norm = []

for i in range(len(values_to_compute)):
    std_norm.append(scipy.stats.norm.pdf(values_to_compute[i], 0, 1))

for i in range(len(values_to_compute_norm)):
    norm.append(scipy.stats.norm.pdf(values_to_compute_norm[i], 1.33,4))

plt.plot(values_to_compute,std_norm);plt.grid(True);plt.title('Standard Normal PDF');plt.xlabel('X');plt.ylabel('Density')
plt.show()

plt.plot(values_to_compute_norm,norm);plt.grid(True);plt.title('Normal PDF @ Mu = 1.33, StdDev = 4');plt.xlabel('X');plt.ylabel('Density')
plt.show()
```

```python
##CDF Standard Normal/Normal
def find_cdf(x,mu,std):
    cdfs = scipy.stats.norm.cdf(x,loc=mu,scale=std)
    return cdfs

print(f'CDF Value Standard Normal F(.5): {find_cdf(.5,0,1)}')
print(f'CDF Value: Mu = 3, Sigma = 1.33 Evaluated at F(3): {find_cdf(3,1.33,4)}')
```

## CDF
The CDF Graph represents the cumulative probabilities of the PDF graph. It is the integrated function of the PDF, therefore any point on the X-Axis has a corresponding Y-Value that represents the cumulative probabilities. If x=0 and y=.50 then it indicates that 50% of the values lie below x=0. Since it is integration of the PDF, I chose to 
plot the CDF using discrete piece-wise sums that would generate the appropriate y-values. The range is 0 to 1 for both Std Normal and Normal CDFs. The CDF evaluated at Mu =1.33 and Sigma =4 represents the std-normal curve shifted by mu and scaled by sigma. The x-domain had to be adjusted as well to fit the full graph ranging from -10 to 10.

```python
##Generate CDF From Cumsum of PDF
## I had to convert the lists into arrays first, cumsum doesn't work on list, didn't want to do a for loop to add values
plt.plot(values_to_compute, np.array(std_norm).cumsum()/10);plt.grid(True);plt.title('CDF Standard Normal Variable');plt.xlabel('X');plt.ylabel('Cumulated Probabilities')
plt.show() #Divide by 10, Cumsum of the values produces a range from 0 to 10 not 0 to 1, however I attributed it to scaling issue

plt.plot(values_to_compute_norm,np.array(norm).cumsum()/10);plt.grid(True);plt.title('CDF Normal Variable Mu = 1.33, Std = 4');plt.xlabel('X');plt.ylabel('Cumulated Probabilities')
plt.show()
```
